```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
with open("features.txt") as f:
    lines = f.readlines()
    # for i in lines:
    #     feature.append(i.replace("\n", "").split()[1])
    features = list(map(lambda x : x.replace("\n", "").split()[1], lines))
features.append("Activity")
print(features)
```

```python
X_train = pd.read_csv("X_train.txt", sep='\s+', header=None)
y_train = pd.read_csv("y_train.txt", sep='\s+', header=None)

train_data = pd.concat([X_train, y_train], axis=1)
train_data.columns = features
train_data.to_csv("train_data.csv", index=False)
```

```python
X_test = pd.read_csv("X_test.txt", sep='\s+', header=None)
y_test = pd.read_csv("y_test.txt", sep='\s+', header=None)
test_data = pd.concat([X_test, y_test], axis=1)
test_data.columns = features
test_data.to_csv("test_data.csv", index=False)
```

```python
df_train = pd.read_csv("train_data.csv")
```

```python
df_test = pd.read_csv("test_data.csv")
```

```python
correlation_matrix = df_train.corr()
redundant_features = set()
threshold = 0.8
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) > threshold:
            colname = correlation_matrix.columns[i]
            redundant_features.add(colname)

groups = {}
for feature in redundant_features:
    for i in range(len(correlation_matrix.columns)):
        if feature != correlation_matrix.columns[i] and abs(correlation_matrix.loc[feature, correlation_matrix.columns[i]]) > threshold:
            if feature not in groups:
                groups[feature] = []
            groups[feature].append(correlation_matrix.columns[i])
```
